```python
records = []
for line in lines:
    if line[0]=="#":
        s = line.split(" ")
        y,d,m,h = s[1], s[2], s[3], s[4]
    else:
        if y in ['2022', '2023']:
            date = " ".join([y,d,m])
            hour = h
            data = [0]*13
            line = re.sub(' +',' ',line)
            #print(line)
            r = line.split(" ")
            #print(r)
            if len(r) <= 8:
                continue
            #break
            data[0], data[1] = r[0][0], r[0][1]
            data[2] = r[1]
            data[3], data[4] = (r[2][:-2], r[2][-1])  if r[2] and r[2][-1] in ['A', 'B'] else (r[2], np.nan)
            data[5], data[6] = (r[3][:-2], r[3][-1])  if r[3] and r[3][-1] in ['A', 'B'] else (r[3], np.nan)
            data[7], data[8] = (r[4][:-2], r[4][-1])  if r[4] and r[4][-1] in ['A', 'B'] else (r[4], np.nan)
            data[9] = r[5] if r[5] else np.nan
            data[10] = r[6] if r[6] else np.nan
            data[11] = r[7] if r[7] else np.nan
            data[12] = r[8] if r[8] else np.nan
            l = []
            l.append(date)
            l.append(hour)
            l.extend(data)
            records.append(l)
```

```python
header = ['Date', 'Hour','LVLTYP1', 'LVLTYP2', 'ETIME', 'PRESS', 'PFLAG', 'GPH', 'ZFLAG', 'TEMP', 'TFLAG', 'RH', 'DPDP', 'WDIR', 'WSPD']
print(len(header))
```

```python
#df = pd.DataFrame([[0]*len(header)]*len(records) , columns= header)
df = pd.DataFrame(records , columns= header)
```

```python
df.head()
```

```python
df.replace('-9999', np.nan, inplace=True)
df.replace('-8888', np.nan, inplace=True)
```

```python
df['RH'] = df['RH'].astype(float)/10.0
```

```python
#pattern = r'^[-]?\d*\.?\d+$'
pattern = r'[^-0-9]'
df['TEMP'] = df['TEMP'].replace({pattern: np.nan}, regex=True)
df['TEMP'].replace('', np.nan, inplace=True)
df['TEMP'] = pd.to_numeric(df['TEMP'], errors='coerce')
```

```python
df['TEMP'] = df['TEMP'].astype(float)/10.0
```

```python
def convert_RH(RH_water, T):
    e_w = 6.112 * math.exp((17.67 * T) / (T + 243.5))
    e_i = 6.112 * math.exp((22.46 * T) / (T + 272))
    RH_ice = (e_w/e_i)*RH_water
    
    return RH_ice
```

```python
df['RH_ice'] = df.apply(lambda row: convert_RH(row['RH'], row['TEMP']), axis=1)
```
